```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import ipywidgets as widgets
from ipywidgets import interact
import scipy.stats as stats
```

```python
df = pd.read_csv('train.csv', sep=',')
```

```python
df = pd.read_csv('train.csv', sep=',')
#df = df[df.LotArea < 30000]
lot = df.LotArea
price = df.SalePrice
```

```python
fig = plt.figure(figsize=(9,5))
ax = fig.add_subplot()

ax.scatter(lot, price, label="LotArea vs SalePrice", color="red", s=2)
ax.set_title("LotArea vs SalePrice")
ax.set_xlabel("LotArea")
ax.set_ylabel("SalePrice")
ax.grid()
ax.legend()
```

```python
deg = 1
data_reg = np.polyfit(lot, price, deg)
xp = np.linspace(0.0, 75000, len(df.LotArea))
yp = np.polyval(data_reg, xp)

fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot()
ax.scatter(lot, price, label="LotArea vs SalePrice", color="gray")
ax.plot(xp, yp, label="Pol. Reg.")
ax.set_title("LotArea vs SalePrice")
ax.grid()
ax.set_xlabel("LotArea")
ax.set_ylabel("SalePrice")
```
