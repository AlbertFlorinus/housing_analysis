```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('house-data/train.csv')
fig = plt.figure(figsize=(9,5))

lotArea = df.LotArea
sales = df.SalePrice

ax = fig.add_subplot()
ax.scatter(sales, lotArea)
ax.set_title('Sales VS Area')
ax.set_xlabel('Sale Price')
ax.set_ylabel('Lot Area')
```
